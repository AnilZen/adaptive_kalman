```python
from classes import *
```

```python
# Transient solution for 3 months
Transient=Lorenz(days=90)
Transient.solve(init_data=np.random.normal(0.25*8,0.5*8,40))
```

## Fig 1

```python
# Nature reference solution for 5 years and 5 days (the three days are for the forecasting system) 
total_days = 5*360
forecast_days = 5
Nature=Lorenz(days=total_days+forecast_days)
Nature.solve(init_data=Transient.sol[-1])
```

```python
stations = np.array([5,15,25,35])
fb=Forecast(total_days,forecast_days,Nature.sol, stations=stations, bias=3., pert_type="Bias")
ns=Forecast(total_days,forecast_days,Nature.sol, stations=stations, bias=0., noise = 2., pert_type="Noise")
gb=Forecast(total_days,forecast_days,Nature.sol, stations=stations, bias=0., noise = 0., pert_type="GridBias")
```

```python
for model in [fb,ns,gb]:
    model.make_forecast()
    model.regression_pars()
```

```python
Nature.solve(init_data=Nature.sol[-1])
```

```python
for model in [fb,ns,gb]:
    model.Truth = Nature.sol
    model.make_forecast()
    model.make_mos_forecast()
    model.error_norms()
```

```python
fig, axs = plt.subplots(1,3, sharey=True, figsize=(15,5))
fig.suptitle("Comparison of Model and MOS forecasts")
for ax in axs.flat:
    ax.set(xlabel='Forecast hour', ylabel='Error')
axs[0].plot(fb.hours,fb.errors, label="Forcing Bias")
axs[0].plot(fb.hours,fb.mos_errors, label="MOS Forecast")
axs[0].legend(loc=2)
axs[1].plot(ns.hours,ns.errors, label="Noise")
axs[1].plot(ns.hours,ns.mos_errors, label="MOS Forecast")
axs[1].legend(loc=2)
axs[2].plot(gb.hours,gb.errors, label="Grid Bias")
axs[2].plot(gb.hours,gb.mos_errors, label="MOS Forecast")
axs[2].legend(loc=2)
for ax in axs.flat:
    ax.label_outer()
plt.savefig("figs/ModelMOS.pdf")
```

## Figure 2

```python
def dep_indep(model):
    model.make_forecast()
    model.make_mos_forecast()
    model.error_norms()
    independent_mos_errors=model.mos_errors.copy()
    model.regression_pars()
    model.make_mos_forecast()
    model.error_norms()
    dependent_mos_errors=model.mos_errors
    return independent_mos_errors-dependent_mos_errors
```

```python
stations = np.array([5,15,25,35])
forecast_days = 5
```

```python
# Nature reference solution for 1 year and 5 days (the additional days are for the forecasting system) 
total_days = 1*360
Nature=Lorenz(days=total_days+forecast_days)
Nature.solve(init_data=Transient.sol[-1])

fb=Forecast(total_days,forecast_days,Nature.sol, stations=stations, bias=3., pert_type="Bias")
ns=Forecast(total_days,forecast_days,Nature.sol, stations=stations, bias=0., noise = 2., pert_type="Noise")
gb=Forecast(total_days,forecast_days,Nature.sol, stations=stations, bias=0., noise = 0., pert_type="GridBias")
```

```python
for model in [fb,ns,gb]:
    model.make_forecast()
    model.regression_pars()

Nature.solve(init_data=Nature.sol[-1])

comps1yr = []
for model in [fb, ns, gb]:
    model.Truth = Nature.sol
    comps1yr.append(dep_indep(model))
```

```python
fig, axs = plt.subplots(1,3, sharey=False, figsize=(15,5))
fig.suptitle("Comparison of dependent and independent forecast accuracy")
for ax in axs.flat:
    ax.set(xlabel='Forecast hour')
axs[0].set(ylabel='Error Norm')
axs[0].plot(fb.hours,comps1yr[0], label="Forcing Bias")
axs[0].plot(fb.hours,comps1yr[1], label="Noise")
axs[0].plot(fb.hours,comps1yr[2], label="Grid Bias")
axs[0].legend(loc=2)
axs[1].plot(fb.hours,comps5yr[0], label="Forcing Bias")
axs[1].plot(fb.hours,comps5yr[1], label="Noise")
axs[1].plot(fb.hours,comps5yr[2], label="Grid Bias")
axs[1].legend(loc=2)
axs[2].plot(fb.hours,comps10yr[0], label="Forcing Bias")
axs[2].plot(fb.hours,comps10yr[1], label="Noise")
axs[2].plot(fb.hours,comps10yr[2], label="Grid Bias")
axs[2].legend(loc=2)
#for ax in axs.flat:
#    ax.label_outer()
plt.savefig("figs/DepIndep.pdf")
```
